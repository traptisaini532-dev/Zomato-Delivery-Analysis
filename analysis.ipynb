# Change appropriate datatypes in the columns.

```python
df['Delivery_person_Age'] = pd.to_numeric(df['Delivery_person_Age'], errors='coerce')
df['Delivery_person_Ratings'] = pd.to_numeric(df['Delivery_person_Ratings'], errors='coerce')
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d-%m-%Y')
```

```python
df['Time_Orderd'] = df['Time_Orderd'].astype(str).str.strip()
df['Time_Orderd'] = pd.to_datetime(df['Time_Orderd'], errors='coerce').dt.time
```

# Plot a graph

```python
# Delivery Time Distribution
plt.figure()
df['Time_taken (min)'].plot(kind='hist')
plt.title('Delivery Time Distribution')
plt.xlabel('Time Taken')
plt.ylabel('Frequency')
plt.show()
```

```python
# Orders by City
plt.figure()
df['City'].value_counts().plot(kind='bar')
plt.title('Orders by City')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Road_traffic_density', y='Time_taken (min)', data=df)
plt.title('Traffic vs Delivery Time')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Weather_conditions', y='Time_taken (min)', data=df)
plt.title('Weather vs Delivery Time')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='Delivery_person_Ratings', y='Time_taken (min)', data=df)
plt.title('Ratings vs Delivery Time')
plt.show()
```
